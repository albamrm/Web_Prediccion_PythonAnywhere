### Modelo Machine Learning:KNN

KNN es un algoritmo sencillo y fcil de entender, lo cual es ideal si ests comenzando en el mundo de los datos. Una de sus ventajas es que no necesita hacer suposiciones complicadas sobre cmo estn distribuidos los datos. Esto significa que puedes usarlo incluso cuando no conoces a fondo la estructura o el patrn de tus datos.

Adems, KNN es muy til cuando tienes que clasificar datos en ms de dos categoras. Por ejemplo, en el caso del dataset de pinginos, donde queremos identificar a qu especie pertenece un pingino (Adelie, Gentoo o Chinstrap), KNN hace un buen trabajo. Lo que hace es ver qu especies estn ms cerca del pingino que queremos clasificar y lo asigna a la especie que sea ms comn entre sus vecinos ms cercanos.

```python
# Convertimos las variables categricas en variables numricas
penguins['species'] = penguins['species'].astype('category').cat.codes
penguins['island'] = penguins['island'].astype('category').cat.codes
penguins['sex'] = penguins['sex'].astype('category').cat.codes
```

```python
# Guardamos el dataset en un archivo CSV
penguins.to_csv('penguins.csv', index=False)
```

```python
# Separamos las caractersticas (X) y la variable objetivo (y)
X = penguins.drop(columns='species')
y = penguins['species']
```

```python
# Escalamos los datos
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Dividimos los datos
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
# Entrenamos el modelo
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)
```

```python
# Evaluamos el modelo
y_pred = knn.predict(X_test)
```

```python
print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

Realizamos un Cross validation para comprobar que no haya sobreajuste en los datos.

```python
cv_scores = cross_val_score(knn, X_scaled, y, cv=5)  
print("Precisin promedio en validacin cruzada:", cv_scores.mean())
print("Desviacin estndar en validacin cruzada:", cv_scores.std())
```

Lo que indica un buen rendimiento y bajo overfitting.
